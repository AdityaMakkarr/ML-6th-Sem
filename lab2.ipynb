```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "adult.csv"
df = pd.read_csv(file_path)

print(df.columns)
```

```python
df.replace("?", np.nan, inplace=True)
print(df.isnull().sum())
```

```python
import matplotlib.pyplot as plt

numeric_columns = ['age', 'fnlwgt', 'education.num', 'capital.gain', 'capital.loss', 'hours.per.week']


for col in numeric_columns:
    plt.figure(figsize=(6, 4))
    plt.hist(df[col], bins=30, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
import seaborn as sns

nominal_columns = ['workclass', 'education', 'marital.status', 'occupation', 
                   'relationship', 'race', 'sex', 'native.country', 'income']

for col in nominal_columns:
    plt.figure(figsize=(8, 4))
    sns.countplot(data=df, y=col, order=df[col].value_counts().index, palette='viridis')
    plt.title(f'Distribution of {col}')
    plt.xlabel('Count')
    plt.ylabel(col)
    plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='age', y='hours.per.week', hue='income', palette='cool', alpha=0.7)
plt.title('Age vs Hours Per Week Comparison')
plt.xlabel('Age')
plt.ylabel('Hours Per Week')
plt.legend(title='Income')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df[:500], x='age', y='hours.per.week', hue='income', palette='cool', alpha=0.7)
plt.title('Age vs Hours Per Week Comparison (First 500 Rows)')
plt.xlabel('Age')
plt.ylabel('Hours Per Week')
plt.legend(title='Income')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
correlation_matrix = df[['age', 'fnlwgt', 'education.num', 'capital.gain', 'capital.loss', 'hours.per.week']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Heatmap of Correlation')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='income', y='age', palette='viridis')
plt.title('Boxplot of Age by Income')
plt.xlabel('Income')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(data=df, x='income', y='age', palette='cool')
plt.title('Violin Plot of Age by Income')
plt.xlabel('Income')
plt.ylabel('Age')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.swarmplot(data=df[:500], x='income', y='age', palette='pastel', size=3)
plt.title('Swarm Plot of Age by Income (First 500 Rows)')
plt.xlabel('Income')
plt.ylabel('Age')
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd


label_encoder = LabelEncoder()
df['sex'] = label_encoder.fit_transform(df['sex'])
df['income'] = label_encoder.fit_transform(df['income'])

df_one_hot = pd.get_dummies(df, columns=['workclass', 'education'], drop_first=True)

print("Label Encoded Columns:")
print(df[['sex', 'income']].head())

print("\nOne-Hot Encoded DataFrame:")
print(df_one_hot.head())
```
